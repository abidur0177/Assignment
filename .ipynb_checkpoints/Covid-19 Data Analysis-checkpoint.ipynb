```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
```

```python
fig,ax = plt.subplots()
index = np.arange(len(data['Country/Region'].head(10)))
width = .3

'''ax.barh(data['Country/Region'].head(10), data['TotalCases'].head(10), width=.4, label='Total Cases')
ax.barh(data['Country/Region'].head(10), data['TotalRecovered'].head(10), width=.4, label='Total Recovered')
ax.barh(data['Country/Region'].head(10), data['TotalDeaths'].head(10), width=.4, label='Total Deaths')'''

ax.bar(index-width, data['TotalCases'].head(10), width=.4, label='Total Cases')
ax.bar(index, data['TotalRecovered'].head(10), width=.4, label='Total Recovered')
ax.bar(index+width, data['TotalDeaths'].head(10), width=.4, label='Total Deaths')

ax.set_title('Covid-19 status of Top 10 Countries')
ax.set_xlabel('Country')
ax.set_ylabel('Numbers')
ax.legend(loc='upper right')


plt.tight_layout()
plt.show()
```
