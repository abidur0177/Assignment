# showing TotalCases, TotalRecovered and TotalDeaths bar plots with respect to Country/Region in logarithmic scale

```python
fig,ax = plt.subplots()
x_index = np.arange(len(data['Country/Region'].head(10)))
width = .3

'''ax.barh(data['Country/Region'].head(10), data['TotalCases'].head(10), width=.4, label='Total Cases')
ax.barh(data['Country/Region'].head(10), data['TotalRecovered'].head(10), width=.4, label='Total Recovered')
ax.barh(data['Country/Region'].head(10), data['TotalDeaths'].head(10), width=.4, label='Total Deaths')'''

ax.bar(x_index-width, data['TotalCases'].head(10), width=.3, color='#2e52a6', label='Total Cases', alpha=.75)
ax.bar(x_index, data['TotalRecovered'].head(10), width=.3, color='#12e66e', label='Total Recovered', alpha=.75)
ax.bar(x_index+width, data['TotalDeaths'].head(10), width=.3, color='#fa232e', label='Total Deaths', alpha=.75)

ax.set_title('Covid-19 status of Top 10 Countries')
ax.set_xlabel('Country')
ax.set_ylabel('Numbers in Log Scale')
ax.legend(loc='upper right')

plt.xticks(ticks=x_index, labels=data['Country/Region'].head(10), rotation=45)
plt.yscale('log')
plt.tight_layout()
plt.show()
```
